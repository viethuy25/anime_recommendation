# Recommendation system

Layer 1

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import KNNImputer

# Assuming data is loaded into the respective dataframes
# Preprocess anime_df

# Selecting relevant numerical columns for now
anime_df_numeric = anime_df[["Score", "Episodes", "Rank", "Popularity", "Favorites", "Scored By", "Members"]]


imputer = KNNImputer(n_neighbors=5)
anime_df_numeric_imputed = imputer.fit_transform(anime_df_numeric)
anime_df_numeric = pd.DataFrame(anime_df_numeric_imputed, columns=anime_df_numeric.columns)

# # Fill NaN values
# anime_df_numeric.fillna(anime_df_numeric.mean(), inplace=True)
```

```python
# Standardize
scaler = StandardScaler()
anime_df_numeric_scaled = scaler.fit_transform(anime_df_numeric)

# One-hot encoding for a sample categorical column, for demonstration
encoder = OneHotEncoder()
genres_encoded = encoder.fit_transform(anime_df["Genres"].str.get_dummies(sep=', '))

# Convert the csr_matrix to a dense DataFrame
genres_encoded_df = pd.DataFrame(genres_encoded.toarray())

# Concatenate the two DataFrames
anime_processed = pd.concat([pd.DataFrame(anime_df_numeric_scaled), genres_encoded_df], axis=1)

# Dimensionality Reduction using Truncated SVD
svd = TruncatedSVD(n_components=15)
anime_reduced = svd.fit_transform(anime_processed)
```

Layer 2

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset

# Encoding users and animes as unique integers
user_score_df['user_index'] = user_score_df['user_id'].astype('category').cat.codes
user_score_df['anime_index'] = user_score_df['anime_id'].astype('category').cat.codes

n_users = user_score_df['user_index'].nunique()
n_animes = user_score_df['anime_index'].nunique()

class RatingDataset(Dataset):
    """PyTorch Dataset for User-Anime Ratings."""
    def __init__(self, users, animes, ratings):
        self.users = users
        self.animes = animes
        self.ratings = ratings

    def __len__(self):
        return len(self.users)

    def __getitem__(self, idx):
        return torch.LongTensor([self.users[idx]]), torch.LongTensor([self.animes[idx]]), self.ratings[idx]

class NCF(nn.Module):
    """Neural Collaborative Filtering model."""
    def __init__(self, n_users, n_animes, n_factors=20, hidden=[10], dropout=0.2):
        super(NCF, self).__init__()
        self.user_factors = nn.Embedding(n_users, n_factors)
        self.anime_factors = nn.Embedding(n_animes, n_factors)
        
        # Construct hidden layers
        self.layers = nn.Sequential(
            *[nn.Sequential(
                nn.Linear((i+1) * n_factors if i == 0 else hidden[i-1], hidden[i]),
                nn.ReLU(),
                nn.Dropout(dropout)
            ) for i in range(len(hidden))]
        )
        self.final = nn.Linear(hidden[-1], 1)
        
    def forward(self, user, anime):
        u = self.user_factors(user)
        a = self.anime_factors(anime)
        x = torch.cat([u, a], dim=1)  # concatenate user and anime embeddings
        x = self.layers(x)
        out = self.final(x)
        return out.squeeze(1)

# Create Dataset and DataLoader
dataset = RatingDataset(user_score_df['user_index'].values, user_score_df['anime_index'].values, user_score_df['rating'].values)
loader = DataLoader(dataset, batch_size=256, shuffle=True)
```

```python
from tqdm import tqdm

# Create Model, Loss, Optimizer
model_ncf = NCF(n_users, n_animes, hidden=[50, 20], dropout=0.2)
loss_function = nn.L1Loss()
optimizer = torch.optim.SGD(model_ncf.parameters(), lr=0.01)

# device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
# model_ncf.to(device)

# Training loop
epochs = 1
for epoch in range(epochs):
    for user, anime, rating in tqdm(loader):
        # Forward pass
        prediction = model_ncf(user, anime)      
        loss = loss_function(prediction.float(), rating.float())
        
        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    
    print(f"Epoch {epoch+1}/{epochs} Loss: {loss.item()}")
```

Layer 3

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split

# Extract the embeddings for users and animes directly from the PyTorch model's embedding layers.
# Assuming the embeddings are the first two layers of your model:
user_embeddings = model_ncf.user_factors.weight.detach().numpy()
anime_embeddings = model_ncf.anime_factors.weight.detach().numpy()

# # Check if we have the same number of embeddings as ratings, if not, you'll need to adjust how you map embeddings to ratings.
# assert user_embeddings.shape[0] == user_score_df['user_id'].nunique()
# assert anime_embeddings.shape[0] == user_score_df['anime_id'].nunique()


# # Map the embeddings to the user_score_df.
# user_factors = user_embeddings[user_score_df['user_id'].values]
# anime_factors = anime_embeddings[user_score_df['anime_id'].values]
valid_user_ids = set(range(user_embeddings.shape[0]))
valid_anime_ids = set(range(anime_embeddings.shape[0]))

user_score_df = user_score_df[user_score_df['user_id'].isin(valid_user_ids) & user_score_df['anime_id'].isin(valid_anime_ids)]

# Now, you can directly use the user_id and anime_id columns as indices
user_factors = user_embeddings[user_score_df['user_id'].values]
anime_factors = anime_embeddings[user_score_df['anime_id'].values]


# Combine user and anime embeddings to serve as input for our NN.
X = np.hstack([user_factors, anime_factors])
y = user_score_df['rating'].values

# Split data into training and test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=69)

# Build the Neural Network.
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dropout(0.2),
#     tf.keras.layers.Dense(256, activation='relu'),
#     tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mae', metrics=['mae', 'accuracy'])

# Train the model.
history = model.fit(X_train, y_train, epochs=3, batch_size=512, validation_data=(X_test, y_test))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Assuming user_embeddings and anime_embeddings are obtained from Layer 3
# Get the user index based on user ID
user_index = user_score_df.iloc[np.where(user_score_df['user_id'].values == user_id)[0]]

# Extract embeddings for the given user index
# user_embeddings[user_score_df['user_id'].values]
user_embedding_selected = user_embeddings[user_index.index]
anime_embeddings_selected = anime_embeddings[user_index.index]

# Calculate the cosine similarity between the user's embedding and all anime embeddings
similarities = cosine_similarity(user_embedding_selected, anime_embeddings_selected)

# Get indices of top N users recommendations (e.g., top 3 most matching users)
num_recommendations = 3
top_indices = similarities.argsort()[-num_recommendations:][::-1]
```

```python
top_3_rating = {}

for row in top_indices:
#     print (row)
    for index in tqdm(row):
        data_selected = user_score_df[user_score_df['anime_index'] == index]['rating']
        top_anime = data_selected.sort_values(ascending = False)[:1]
        title_name = user_score_df[user_score_df['anime_index'] == index]['Anime Title'][:1].values[0]
        
        if title_name not in top_3_rating.keys():
            top_3_rating[title_name] = top_anime.values[0]

top_3_rating
```
